## An치lisis descriptivo de los datos

```python
import pandas as pd
import numpy as np

#Fijar la semilla aleatoria
np.random.seed(42)


data = pd.read_excel('DataBaseTFM.xlsx')
etiq_verd=data[['Etiq']]
data=data[['Temperatura','Consumo_Lavanderia', 'Consumo_Cocina', 
           'Resto_Consumo','Habitantes']]

# An치lisis descriptivo
descripcion = data.describe()

# Imprimir el an치lisis descriptivo
print(descripcion)
```

## Normalizaci칩n de los datos

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data_normalized = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
```

## Histogramas de las variables

```python
import matplotlib.pyplot as plt

# N칰mero de columnas para los histogramas
num_columns = len(data_normalized.columns)
num_rows = num_columns // 3 + (num_columns % 3 > 0)  # Asegurar que haya suficientes filas

# Crear una figura con subplots
fig, axs = plt.subplots(num_rows, 3, figsize=(15, num_rows * 5))  # Ajustar el tama침o seg칰n sea necesario

# Aplanar el arreglo de ejes para facilitar su manejo
axs = axs.flatten()

# Crear un histograma para cada columna normalizada
for i, column in enumerate(data_normalized.columns):
    axs[i].hist(data_normalized[column])
    axs[i].set_title(f'({chr(65 + i)}): {column}')
    axs[i].set_xlabel('Valor')
    axs[i].set_ylabel('Frecuencia')
    
# Ocultar los ejes adicionales si los hay
for ax in axs[num_columns:]:
    ax.set_visible(False)
    
# Ajustar la separaci칩n entre las filas de subplots
plt.subplots_adjust(hspace=.01)  # Ajusta el valor seg칰n sea necesario
    
# Ajustar el layout y mostrar la figura
plt.tight_layout()
plt.show()
```

## Matriz de correlaci칩n

```python
matriz_corr=data_normalized.corr(method='spearman') #Coeficiente de Spearman
print(matriz_corr)
```

### Mapa de calor de la matriz de correlaci칩n

```python
import seaborn as sns
# Crear mapa de calor
mapa_calor = sns.heatmap(matriz_corr, 
                        annot=True,  # Anotar valores en las celdas
                        cmap="coolwarm")  # Paleta de colores (personalizable)
plt.show()
```

## Divisi칩n de los datos en data_train y data_val

```python
from sklearn.model_selection import train_test_split

data=data_normalized
data['Etiq'] = etiq_verd['Etiq']
print(data.head())
data_train, data_test = train_test_split(data, test_size=0.2, random_state=51)

print('\n ------------------------ Datos de entrenamiento ------------------------------ \n')
print(data_train.head())
print('\n ------------------------ Datos de validaci칩n ---------------------------------- \n')
print(data_test.head()) 

index_data_train=data_train.index
index_data_test=data_test.index

# Guardar los datos en archivos excel
data_train.to_excel('Datos_train.xlsx')
data_test.to_excel('Datos_test.xlsx')

data_train=np.array(data_train) 
data_test=np.array(data_test) 

Anom_reales=[2,   3,   4 ,  5 , 24,  27, 152, 209, 217, 221, 222, 228, 247, 
             248, 252, 254, 299, 300, 530, 612 ,613, 618, 660, 662,666, 681, 709, 713]

anom_data_train= np.intersect1d(index_data_train, Anom_reales)
anom_data_test= np.intersect1d(index_data_test, Anom_reales)
print('\n -------------------------------------------------------------------------------\n ')
print("\n Cantidad de anomal칤as en data_train:",len(anom_data_train))
print(anom_data_train)
print("\n Cantidad de anomal칤as en data_test:",len(anom_data_test))
print(anom_data_test)
```

## Histogramas de los datos de entrenamiento y los de validaci칩n

```python
data_train=pd.read_excel('Datos_train.xlsx',index_col=0)
data_test = pd.read_excel('Datos_test.xlsx',index_col=0)  
etiq_verd_train=data_train[['Etiq']]
etiq_verd_test=data_test[['Etiq']]
data_train = data_train[['Temperatura','Consumo_Lavanderia', 'Consumo_Cocina', 
                         'Resto_Consumo', 'Habitantes']]
data_test = data_test[['Temperatura','Consumo_Lavanderia', 'Consumo_Cocina', 
                         'Resto_Consumo', 'Habitantes']]

fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(12, 20))
# Iterar sobre las columnas de los dataframes
for i, column in enumerate(data_train.columns):
    # Histograma para data_train
    axes[i, 0].hist(data_train[column], bins=20, alpha=0.7, color='blue')
    axes[i, 0].set_title(f'Conjunto de entrenamiento - {column}')
    
    # Histograma para data_test
    axes[i, 1].hist(data_test[column], bins=20, alpha=0.7, color='green')
    axes[i, 1].set_title(f'Conjunto de validaci칩n - {column}')

# Ajustar el layout para que no se solapen los t칤tulos
plt.tight_layout()
plt.show()
```
